```python
from datetime import date
from datetime import datetime, date
dF['Date.of.Birth'] = pd.to_datetime(dF['Date.of.Birth'])
dF['born'] = dF['Date.of.Birth'].dt.strftime('%y').astype(int)
dF['current_year'] = 22
dF['age'] = dF['born'] - dF['current_year']
```

```python
# Creating another column on the base of age column and caterorize them for adult and old  
dF['Person'] = dF['age'].apply(lambda x : 'old' if x >= 60 else 'adult' )

```

```python
# # Find the average salaries of adult and old people in dataset
dF.groupby('Person')['disbursed_amount'].mean()
```
